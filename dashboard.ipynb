```python
# dashboard.py
import ipywidgets as widgets
from IPython.display import display
from service import *

output = widgets.Output()

init_registry()

def build_ui():

    csv_input = widgets.Text(description="CSV")
    raw_input = widgets.Text(description="RAW")
    clean_input = widgets.Text(description="CLEAN")

    table_dropdown = widgets.Dropdown(options=list_tables(), description="表")

    btn_create = widgets.Button(description="创建", button_style="success")
    btn_delete = widgets.Button(description="删除", button_style="danger")
    btn_view = widgets.Button(description="查看")
    btn_dropall = widgets.Button(description="删除所有表", button_style="warning")

    # -------------------------
    def on_create(_):
        with output:
            output.clear_output()

            try:
                r, c = create_dataset(
                    csv_input.value.strip(),
                    raw_input.value.strip(),
                    clean_input.value.strip(),
                    )

                table_dropdown.options = list_tables()

                print(f"✔ 创建成功 raw={r} clean={c}")

            except Exception as e:
                print("❌", e)

    def on_delete(_):
        with output:
            output.clear_output()

            delete_dataset(table_dropdown.value)
            table_dropdown.options = list_tables()

            print("✔ 删除成功")

    def on_view(_):
        with output:
            output.clear_output()
            df = load_table(table_dropdown.value)
            display(df.head())

    def on_dropall(_):
        with output:
            output.clear_output()
            drop_all_tables()
            table_dropdown.options = list_tables()
            print("✔ 已删除所有表")

    btn_create.on_click(on_create)
    btn_delete.on_click(on_delete)
    btn_view.on_click(on_view)
    btn_dropall.on_click(on_dropall)

    ui = widgets.VBox([
        csv_input,
        raw_input,
        clean_input,
        widgets.HBox([btn_create, btn_delete, btn_view, btn_dropall]),
        table_dropdown,
        output
    ])

    display(ui)

build_ui()
```
